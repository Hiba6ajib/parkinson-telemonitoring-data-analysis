```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn as sk
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
model=LinearRegression(fit_intercept=True)
```

```python
data=pd.read_csv("parkinsons.data")
```

```python
data.describe()
```

```python
#NHR= noise to harmonic ratio
#HNR= harmonic to noise ratio
x=data.NHR
y=data.HNR
plt.title("NHR and HNR")
plt.xlabel('Noise to Harmonic Ratio')
plt.ylabel('Harmonic to Noise Ratio')
plt.plot(x,y,'o',color='#00BFFF')
```

```python
#unified parkinsons disease rating scale(UPDRS)
x=data.motor_UPDRS
y=data.total_UPDRS
params=np.polyfit(x,y,2)
xp=np.linspace(x.min(),x.max(),20)
yp=np.polyval(params,xp)
plt.plot(xp,yp,'k',alpha=0.8,linewidth=1)
plt.plot(data.motor_UPDRS,data.total_UPDRS,'o',markersize=5,color='#BF3EFF')
sig=np.std(y-np.polyval(params, x))
plt.fill_between(xp,yp-sig,yp+sig,
                color='k',alpha=0.2)
plt.title('Motor and Total of Unified Parkinsons Disease Rating Scale')
plt.xlabel('Motor UPDRS')
plt.ylabel('Total UPDRS')
```

```python
#RPDE= recurrence period density entropy
#PPE=Pitch Period Entropy
[data['RPDE'].hist(color='#CD8C95'),data['PPE'].hist(color='#98F5FF')]
plt.title("Recurrence Period Density Entropy vs Pitch Period Entropy")
```

```python
a=data[['Shimmer']].values
b=data['Jitter_Percent'].values
a.shape
```

```python
model.fit(a,b)
r_line= lambda a: model.intercept_ + model.coef_ * a
```

```python
fig ,ax=plt.subplots(1,1,figsize=(10,10))
a_val=np.linspace(0,0.5,2)
ax.plot(a_val,r_line(a_val),color='red',linewidth=1,label="Regression Line")
ax.scatter(a,b,color='green',alpha=0.2,label='data')

plt.xlabel('Shimmer')
plt.ylabel('Jitter Percent')
plt.title('Comparison of Shimmer vs Jitter')
```
